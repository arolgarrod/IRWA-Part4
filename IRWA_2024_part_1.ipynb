### Dictionary with original tweets

As a first step, we are creating a dictionary with all tweet important information: <br>
- tweet
- hashtags
- url
- date
- likes
- retweets

In this dictionary, we are storing the original tweet, without any pre-processing.

```python
def extract_hashtags(line):
    """
    Description:
    Extracts hashtags from a line of text and returns the list of hashtags.

    Input:
    - line (str): A string containing the line of text, which may include hashtags.

    Output:
    - hashtags (list of str): A list containing all hashtags found in the line.
    """
    hashtags = re.findall(r'#\w+', line)

    # Return tweet hashtags
    return hashtags
```

### Dictionary with processed tweets

As a next step, we are preprocessing the text as follows: <br>
- lower case
- extracting hashtags (we are saving it into another field as we want to differentiate them from the rest)
- Remove any symbol that does not match with a to z (character), 0 to 9 (digit) and white space character
- Remove stopwords
- Stemming
- URL removal (just for tweet content)<br>

‚ÄºÔ∏è The above process is done with tweet itself and hashtags separately, so that hashtags words are preprocessed but not included into the tweet

```python
def build_terms(line):
    """
    Description:
    Preprocess the text (tweet) by removing stop words, punctuation (but keeping numbers), URLs,
    stemming, transforming to lowercase, and extracting hashtags. The hashtags are excluded
    from the preprocessed text.

    Input:
    line -- string (text) to be preprocessed

    Output:
    preprocessed_line -- a list of tokens corresponding to the input text after preprocessing
    hashtags -- a list of extracted hashtags
    """

    stemmer = PorterStemmer()
    stop_words = set(stopwords.words("english"))

    # Transform to lowercase
    line = line.lower()

    # HASHTAGS

    # Extract hashtags (keeping the # symbol)
    hashtags = re.findall(r'#\w+', line)
    # Remove punctuation but keep numbers
    hashtags = [re.sub(r'[^a-z0-9\s]', '', word) for word in hashtags]
    # Remove stopwords
    hashtags = [word for word in hashtags if word not in stop_words]
    # Perform stemming
    hashtags = [stemmer.stem(word) for word in hashtags]
    # Remove empty strings (in case stemming or other operations result in an empty hashtag)
    hashtags = [word for word in hashtags if word]

    # LINE

    # Remove emojis using demoji
    line = demoji.replace(line, " ")
    # Remove hashtags from the text --> remove all the word
    line = re.sub(r'#\w+', '', line)
    # Remove URLs
    line = re.sub(r'http\S+|www\S+|https\S+', '', line)
    # Remove punctuation but keep numbers (and hashtags are already removed)
    line = re.sub(r'[^a-z0-9\s]', '', line)
    # Tokenize the text to get a list of terms
    line = line.split()
    # Remove the word amp since & it is parsed as &amp and it is one of the most frequent words!
    line = [word for word in line if word != 'amp']
    # Remove stopwords
    line = [word for word in line if word not in stop_words]
    # Perform stemming
    line = [stemmer.stem(word) for word in line]

    # Return preprocessed text and hashtags
    return line, hashtags # return the preprocessed tweet content and hashtags
```

Finally, let's create a dictionary to store the preprocessed tweets contents and hashtags:

```python
processed_tweets_dict = {}

for i in range(len(farmers_protest_tweets)):
    try:
        id_num = farmers_protest_tweets[i]['id']
        doc_id = map_ids.loc[map_ids['id'] == id_num, 'docId'].values[0]

        line = farmers_protest_tweets[i]['content']

        tweet, hashtags = build_terms(line)

        processed_tweets_dict[doc_id] = {
            'tweet': tweet,
            'hashtags': hashtags,
        }
    except:
        continue
```

# Part 2: Exploratory Data Analysis

All the results are shown in the report.

```python
import spacy
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
```

```python
# Initialize Spacy for Named Entity Recognition (NER)
nlp = spacy.load("en_core_web_sm")

# Prepare data for EDA
original_tweets_df = pd.DataFrame.from_dict(original_tweets_dict, orient='index')
processed_tweets_df = pd.DataFrame.from_dict(processed_tweets_dict, orient='index')

# 1. Word Count Distribution
def word_count_distribution(original_df,processed_df,bins=30):
  """
    Description:
    This function calculates the word count of each tweet from a processed DataFrame, stores it in the original DataFrame,
    and plots the word count distribution using a histogram.

    Input:
    - original_df (DataFrame): The original DataFrame where the word count column will be added.
    - processed_df (DataFrame): The DataFrame containing the 'tweet' column (processed) used to compute the word count.
    - bins (int, optional): The number of bins to use in the histogram. Default is 30.

    Output:
    - A histogram showing the distribution of word counts in the tweets with a kernel density estimate (KDE).
    - The original DataFrame will have a new column called 'word_count' with the word count for each tweet.
    """

  original_df['word_count'] = processed_df['tweet'].apply(lambda x: len(' '.join(x).split()))

  plt.figure(figsize=(10, 6))
  sns.histplot(original_df['word_count'], bins=bins, kde=True)
  plt.title('Word Count Distribution in Tweets')
  plt.xlabel('Word Count')
  plt.ylabel('Frequency')
  plt.show()

# 2. Average Sentence Length
def average_sentence_length(dataset=original_tweets_df):
    """
      Description:
      This function calculates and prints the average sentence length (in words) from the 'word_count' column of the
      input DataFrame.

      Input:
      - Dataset: A Dataframe with 'word_count' column. Original_tweets_df' as default.

      Output:
      - Prints the average sentence length formatted to two decimal places.
      """
    avg_length = dataset['word_count'].mean()
    print(f'Average Sentence Length: {avg_length:.2f} words')

# 3. Vocabulary Size
def vocabulary_size(dataset=processed_tweets_df):
    """
      Description:
      This function calculates and prints the vocabulary size, which is the number of unique words found in the 'tweet' column
      of the given dataset. It assumes the 'tweet' column contains pre-processed tweets in list format.

      Input:
      - dataset (DataFrame): The DataFrame containing a 'tweet' column. Default is 'processed_tweets_df'.

      Output:
      - Prints the vocabulary size, which represents the count of unique words in the dataset.
      """
    all_words = [word for tweet in dataset['tweet'] for word in tweet]
    vocab_size = len(set(all_words))
    print(f'Vocabulary Size: {vocab_size} unique words')

# 4. Ranking of Tweets by Retweets
def ranking_by_retweets(dataset=original_tweets_df):
    """
      Description:
      This function identifies and prints the top 10 tweets with the highest number of retweets from the given dataset.
      It assumes that the dataset contains a 'retweets' column along with the 'tweet' column.

      Input:
      - dataset (DataFrame): The DataFrame containing 'tweet' and 'retweets' columns. Default is 'original_tweets_df'.

      Output:
      - Prints the top 10 tweets ranked by the number of retweets.
      """
    top_retweets = dataset[['tweet', 'retweets']].sort_values(by='retweets', ascending=False).head(10)
    print('Top 10 Tweets by Retweets:')
    print(top_retweets)

# 5. Word Cloud for Most Frequent Words
def generate_wordcloud(dataset=processed_tweets_df):
    """
      Description:
      This function generates and displays a word cloud from the words in the 'tweet' column of the given dataset.
      It assumes that the tweets are represented as lists of words.

      Input:
      - dataset (DataFrame): The DataFrame containing a 'tweet' column. Default is 'processed_tweets_df'.

      Output:
      - Displays a word cloud visualization of the most frequent words in the tweets.
      """
    # Flatten the list of lists into a single list of words
    all_words = [word for tweet in dataset['tweet'] for word in tweet]

    # Join the words into a single string
    all_words_string = ' '.join(all_words)

    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_words_string)

    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# 6. Top n Most Frequent Words
def most_frequent_words(dataset=processed_tweets_df,n=50):
    """
      Description:
      This function calculates and displays the top n most frequent words from the 'tweet' column of the given dataset.
      It assumes that the tweets are represented as lists of words.

      Input:
      - dataset (DataFrame): The DataFrame containing a 'tweet' column. Default is 'processed_tweets_df'.
      - n (int): The number of most frequent words to display. Default is 50.

      Output:
      - Prints the top n most frequent words and their counts.
      - Displays a bar plot of the most frequent words.
      """
    # Flatten the list of lists into a single list of words
    all_words = [word for tweet in dataset['tweet'] for word in tweet]

    # Use Counter to count the frequency of each word
    word_counts = Counter(all_words)

    # Get the top n most frequent words
    most_common_words = word_counts.most_common(n)

    # Convert the result into a DataFrame for easier viewing
    word_freq_df = pd.DataFrame(most_common_words, columns=['Word', 'Count'])

    print(f'Top {n} Most Frequent Words:')
    print(word_freq_df)

    # Plot the results
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Count', y='Word', data=word_freq_df)
    plt.title(f'Top {n} Most Frequent Words')
    plt.show()

# 7. Named Entity Recognition (NER)
def entity_recognition(dataset=processed_tweets_df):

    """
      Description:
      This function performs Named Entity Recognition (NER) on the tweets in the given dataset using spaCy.
      It extracts entities from each tweet and displays the counts of each entity type.

      Input:
      - dataset (DataFrame): The DataFrame containing a 'tweet' column with tweets represented as lists of words.
        Default is 'processed_tweets_df'.

      Output:
      - Prints the counts of entities by type.
      - Displays a count plot of the recognized entities by type.
      """
    entities = []
    for tweet in dataset['tweet']:
        # Join the list of words into a single string
        tweet_text = ' '.join(tweet)
        doc = nlp(tweet_text)  # Process the tweet text with Spacy
        for ent in doc.ents:
            entities.append((ent.text, ent.label_))

    entity_df = pd.DataFrame(entities, columns=['Entity', 'Label'])
    entity_counts = entity_df['Label'].value_counts()

    print('Entity Counts by Type:')
    print(entity_counts)

    plt.figure(figsize=(10, 6))
    sns.countplot(y='Label', data=entity_df, order=entity_counts.index)
    plt.title('Entity Recognition in Tweets')
    plt.show()

# 8. Hashtag Count Distribution
def hashtag_count_distribution(dataset=processed_tweets_df):
    """
      Description:
      This function calculates the distribution of hashtag counts in the given dataset and displays it using a histogram.
      It assumes that the dataset contains a 'hashtags' column, which is a list of hashtags for each tweet.

      Input:
      - dataset (DataFrame): The DataFrame containing a 'hashtags' column. Default is 'processed_tweets_df'.

      Output:
      - Displays a histogram of the distribution of hashtag counts in tweets.
      """
    dataset['hashtag_count'] = dataset['hashtags'].apply(lambda x: len(x))

    plt.figure(figsize=(10, 6))
    sns.histplot(dataset['hashtag_count'], bins=25, kde=False)
    plt.title('Hashtag Count Distribution in Tweets')
    plt.xlabel('Number of Hashtags')
    plt.ylabel('Frequency')
    plt.show()

# 9. Most Frequent Hashtags
def most_frequent_hashtags(dataset=processed_tweets_df,n=10):
    """
      Description:
      This function calculates and displays the top n most frequent hashtags from the 'hashtags' column of the given dataset.
      It assumes that the hashtags are represented as lists within the column.

      Input:
      - dataset (DataFrame): The DataFrame containing a 'hashtags' column with lists of hashtags. Default is 'processed_tweets_df'.
      - n (int): The number of most frequent hashtags to display. Default is 10.

      Output:
      - Prints the top n most frequent hashtags and their counts.
      - Displays a bar plot of the most frequent hashtags.
      """
    # Flatten the list of hashtags into a single list
    all_hashtags = [hashtag for hashtags in dataset['hashtags'] for hashtag in hashtags]

    # Use Counter to count the frequency of each hashtag
    hashtag_counts = Counter(all_hashtags)

    # Get the top n most frequent hashtags
    most_common_hashtags = hashtag_counts.most_common(n)

    # Convert the result into a DataFrame for easier viewing
    hashtag_freq_df = pd.DataFrame(most_common_hashtags, columns=['Hashtag', 'Count'])

    print(f'Top {n} Most Frequent Hashtags:')
    print(hashtag_freq_df)

    # Plot the results
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Count', y='Hashtag', data=hashtag_freq_df)
    plt.title(f'Top {n} Most Frequent Hashtags')
    plt.show()

# 10. Word Cloud for Hashtags
def generate_hashtag_wordcloud(dataset=processed_tweets_df):
    """
      Description:
      This function generates and displays a word cloud from the hashtags in the 'hashtags' column of the given dataset.
      It assumes that the hashtags are represented as lists within the column.

      Input:
      - dataset (DataFrame): The DataFrame containing a 'hashtags' column with lists of hashtags. Default is 'processed_tweets_df'.

      Output:
      - Displays a word cloud visualization of the most frequent hashtags in the dataset.
      """
    # Flatten the list of hashtags into a single list
    all_hashtags = [hashtag for hashtags in dataset['hashtags'] for hashtag in hashtags]

    # Join the hashtags into a single string
    hashtag_string = ' '.join(all_hashtags)

    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(hashtag_string)

    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Hashtag Word Cloud')
    plt.show()
```
